```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('Life_Expectancy_Data.csv')
originaldataset = pd.read_csv('Life_Expectancy_Data.csv') #library importing and getting datasets
```

```python
dataset.columns #printing dataset's columns
```

```python
originaldataset = originaldataset[['Country', 'Year', 'Status', 'Life_expectancy', 'Adult Mortality',
       'infant deaths', 'Alcohol', 'percentage_expenditure', 'Hepatitis B',
       'Measles ', ' BMI ', 'under-five deaths ', 'Polio', 'Total expenditure',
       'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population',
       ' thinness  1-19 years', ' thinness 5-9 years',
       'Income_composition_of_resources', 'Schooling']]
```

```python
dataset = dataset[['Country', 'Year', 'Status', 'Life_expectancy', 'Adult Mortality',
       'infant deaths', 'Alcohol', 'percentage_expenditure', 'Hepatitis B',
       'Measles ', ' BMI ', 'under-five deaths ', 'Polio', 'Total expenditure',
       'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population',
       ' thinness  1-19 years', ' thinness 5-9 years',
       'Income_composition_of_resources', 'Schooling']]
#extracting important variables from dataset.
```

```python
dataset #printing the dataset
```

```python
dataset.dropna(subset = ['Country', 'Year', 'Status', 'Life_expectancy', 'Adult Mortality',
       'infant deaths', 'Alcohol', 'percentage_expenditure', 'Hepatitis B',
       'Measles ', ' BMI ', 'under-five deaths ', 'Polio', 'Total expenditure',
       'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population',
       ' thinness  1-19 years', ' thinness 5-9 years',
       'Income_composition_of_resources', 'Schooling'], inplace=True)
#Deleting the "null" values from dataset.
```

```python
dataset.isna().sum() #summary of na values
```

```python
df=(dataset[['Country', 'Year', 'Status', 'Life_expectancy', 'Adult Mortality',
       'infant deaths', 'Alcohol', 'percentage_expenditure', 'Hepatitis B',
       'Measles ', ' BMI ', 'under-five deaths ', 'Polio', 'Total expenditure',
       'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population',
       ' thinness  1-19 years', ' thinness 5-9 years',
       'Income_composition_of_resources', 'Schooling']] != 0).all(axis=1)
#setting all 0 values into df array.
```

```python
df # printing df array
```

```python
dataset = dataset.loc[df] #deleting the df array from dataset so 0 values are deleted.
```

```python
dataset #printing dataset.
```

```python
from sklearn.preprocessing import OneHotEncoder #Importing OneHotEncoder Library
enc = OneHotEncoder(handle_unknown='ignore')

enc_df = pd.DataFrame(enc.fit_transform(originaldataset[['Country']]).toarray()) #fitting country column into encoder array
dataset = dataset.join(enc_df)  #joining encoder array into dataset
dataset
```

```python
dataset = dataset.drop(['Country','Status'], axis=1) #deleting the string country and status array
dataset
```

```python
x = dataset.drop(['Life_expectancy'],axis = 1) #x values that will predict y.
y = dataset.Life_expectancy #y value that will be predicted with x values.
```

```python
import seaborn as sns #importing seaborn library
import warnings #importing warnings
warnings.filterwarnings('ignore') #ignoring warnings

features_number = dataset.columns[dataset.dtypes != "object"]
for i in features_number:
    plt.figure()
    sns.distplot(dataset[i]) #show the density of every element graphical.
```

```python
from sklearn.model_selection import train_test_split #importing train_test split.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.4, random_state = 8, shuffle = True)
#defining the train and test values and how it will be executed.
```

```python
from sklearn.linear_model import LinearRegression #importing LineerRegression.
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
reg.score(x_train,y_train) #printing the score of our lineer regression models (train).
```

```python
reg.score(x_test,y_test) #printing the score of our lineer regression models (test).
```

```python
# reg.coef_ #this is coefficients. This means like Alcohol has -1.79 effect on Life Expectancy.
```

```python
reg.intercept_ #this also called as constant (line).
```

```python
predictions = reg.predict(x_train)
#setting predictions array to predicted values with x_train.
```

```python
# predictions # printing predictions.
```

```python
from sklearn.metrics import mean_squared_error #importing MSE.

mean_squared_error(y_train,predictions) #calculating and printing MSE
```
